We need the grade data to be converted to wide-form for compatibility with weight matrix constructors later. By 'wide-form,' we mean that the data are structured with the student groups along the columns and the districts as the index.

This notebook will perform this operation through a `pandas.pivot`

```python
import pandas as pd
import geopandas as gpd 
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_column', 200)
```

```python
math = pd.read_csv("clean_data/math.csv")
ela = pd.read_csv("clean_data/ela.csv")
```

```python
ela = ela.rename(columns={'currstatus': 'e_curr', 
                          'priorstatus': 'e_prio', 
                          'change' : 'e_chan'})
math = math.rename(columns={'currstatus': 'm_curr', 
                            'priorstatus': 'm_prio', 
                            'change' : 'm_chan'})

# renaming columns to be more concise
# geopandas limits the amount of characters on 
# collumn names upon reading file - being proactive here
```

```python
math
```

```python
math.drop('Unnamed: 0', axis=1, inplace=True)
ela.drop('Unnamed: 0', axis=1, inplace=True)
# drop old index
```

```python
m = math.pivot_table(index='District Code', columns='studentgroup')
# pandas.pivot_table
```

```python
m
```

```python
m.columns = [' '.join(col).strip() for col in m.columns.values]
# squishing multi-index columns together
```

```python
m
```

```python
m.iloc[2]
# Alamdeda unified
```

```python
e = ela.pivot_table(index='District Code', columns='studentgroup')
# pandas.pivot_table
```

```python
e.columns = [' '.join(col).strip() for col in e.columns.values]
# squishing multi-index columns together
```

```python
e
```

We lose `districtname`, even though I did not explicitly drop it. This shouldn't present an issue later, we'll be merging by the index, `District Code`.

```python
grades = e.merge(m, on=['District Code'])
grades
# cautiously optimistic?
```
