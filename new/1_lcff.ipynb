select desired columns, clean up names

```python
lcff = df[['CountyCode', 
           'DistrictCode', 
           'SchoolCode', 
           'LocalEducationalAgency', 
           'UnduplicatedPupilPercentage\r\nTarget,D-1', 
           'BaseGrantFunding\r\nTarget,C-5', 
           'SupplementalGrantFunding\r\nTarget,D-7', 
           'ConcentrationGrantFunding\r\nTarget,E-8']]
```

```python
lcff = lcff.rename(columns={'CountyCode':'county_fips',
                            'DistrictCode':'district_code',
                            'SchoolCode':'school_code',
                            'LocalEducationalAgency':'name',
                            'UnduplicatedPupilPercentage\r\nTarget,D-1':'unduplicated_pupil_count',
                            'BaseGrantFunding\r\nTarget,C-5':'base_grant',
                            'SupplementalGrantFunding\r\nTarget,D-7':'supplemental_grant',
                            'ConcentrationGrantFunding\r\nTarget,E-8':'concentration_grant'})
```

```python
lcff['district_code'] = lcff['district_code'].astype(float).astype(int).astype(str)
lcff['school_code'] = lcff['school_code'].astype(float).astype(int).astype(str)
```

```python
lcff.base_grant = lcff.base_grant.str.replace(' ', '')
lcff.base_grant = lcff.base_grant.str.replace('$', '')
lcff.base_grant = lcff.base_grant.str.replace(',', '')
lcff.base_grant = lcff.base_grant.str.replace('-', '')
```

```python
lcff.supplemental_grant = lcff.supplemental_grant.str.replace('$', '').str.strip()
lcff.concentration_grant = lcff.concentration_grant.str.replace('$', '').str.strip()
```
